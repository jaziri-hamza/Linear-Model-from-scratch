###### Import Libraris

````python
import matplotlib.pyplot as plt
import numpy as np
from sys import path
path.append('..')
from mylib.models import LinearRegression
````

###### Initisialize some fake data

````python
X = np.linspace(10, 100, 10).reshape(-1, 1)
Y = np.linspace(10, 100, 10).reshape(-1, 1)
````

##### Model 1
Simple linearRegression Model

````python
lin_reg = LinearRegression(X,Y)
lin_reg.fit(history=True)
````

###### Define some functions to plot our data and the Error function

````python
def plotData(X,Y, X_test, Y_pred, name):
    subplot = plt.subplot()
    subplot.set_title(name)
    subplot.set_xlabel('X')
    subplot.set_ylabel('Y')
    subplot.scatter(X,Y, c="red", marker=".")
    subplot.plot(X_test, Y_pred)
    plt.show()
    
def plotCostFn(costs, name="Cost Function"):
    subplot = plt.subplot()
    subplot.set_title(name)
    subplot.set_xlabel('Iteration')
    subplot.set_ylabel('Error Value')
    subplot.plot(costs)
    plt.show()
````

###### Plot data

````python
plotData(X,Y, X, lin_reg.predict(X), "Data")
plotCostFn(lin_reg.history)
````

##### Model 2
Example of Linear Regression Model with **Lasso** Regularization and **Adam** Optimizer

````python
# init the regularization
from mylib.regularizations import LassoRegularization
lasso_reg = LassoRegularization(alpha=0.9)
# init the optimizer
from mylib.optimizers import AdamOptimizer
adam_opt = AdamOptimizer(learning_rate=0.001, alpha=0.5, beta=0.5)
# the model
lin_reg = LinearRegression(X,Y)
lin_reg.fit(iter=2000, optimizer=adam_opt, regularization=lasso_reg, history=True)
````

````python
plotData(X,Y, X, lin_reg.predict(X), "Data")
plotCostFn(lin_reg.history)
````
